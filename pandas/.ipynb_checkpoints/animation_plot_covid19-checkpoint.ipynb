```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

# Get Data

```python
url = "https://covid19.th-stat.com/api/open/cases"
covid_data = requests.get(url).json()["Data"]
covid_df = pd.DataFrame(covid_data)
covid_df.head(5)
```

```python
covid_df.dtypes
```

# Prepare Data

```python
# change type object to datetime using "pd.to_datetime"
covid_df['datetime'] = pd.to_datetime(covid_df['ConfirmDate'])
covid_df.dtypes
```

```python
# Select only the columns that are used
province_df = covid_df.loc[:, ["datetime", "ProvinceEn"]]
province_df["Confirmed"] = 1
province_df.head(5)
```

```python
# sort by datetime
province_sort_df = province_df.sort_values(by="datetime", ascending=True).reset_index(drop=True)
province_sort_df.head(10)
```

```python
# pivot and add datetime column
province_pivot_df = province_sort_df.pivot(columns='ProvinceEn', values='Confirmed').fillna(0)
province_pivot_df['datetime'] = province_sort_df['datetime']
province_pivot_df.head(5)
```

```python
# group by datetime
province_group_df = province_pivot_df.groupby("datetime").sum()
province_group_df.head(10)
```

```python
# cumulative frequency
result_df = province_group_df.cumsum()
result_df.head(10)
```

# Analyze Data

```python
# top 5 province covid confirmed 
last_day = -1
s_sort = result_df.iloc[last_day].sort_values(ascending=False)[:5]
s_sort
```

# Visualize Data

```python
fig, ax = plt.subplots(figsize=(4, 2.5), dpi=144)
colors = plt.cm.Dark2(range(5))
y = s_sort.index
width = s_sort.values
ax.barh(y=y, width=width, color=colors);
for i, v in enumerate(width):
    ax.text(5, i-0.2, str(int(v)), color='black')
```

```python
from matplotlib.animation import FuncAnimation

num_top = 5
labelsize = 4
def init():
    ax.clear()
    nice_axes(ax)
    ax.set_ylim(.2, 6.8)

def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=labelsize, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]

def update(i):
    ax.clear()
    for bar in ax.containers:
        bar.remove()
        
    s_sort = result_df.iloc[i].sort_values(ascending=False)[:num_top]
    y = list(range(5,0,-1))
    width = s_sort.values
    colors = plt.cm.Dark2(range(num_top))
    labels = s_sort.index
    ax.barh(y=y, width=width, color=colors, tick_label=labels)
    for i, v in zip(y, width):
        ax.text(0, i-0.2, str(int(v)), color='black')
    date_str = s_sort.name.strftime("%d/%m/%Y")
    ax.set_title(f'Thai Province COVID-19 Confirmed - {date_str}', fontsize='smaller')
    
fig = plt.Figure(figsize=(4, 2.5), dpi=144)
ax = fig.add_subplot()
anim = FuncAnimation(fig=fig, func=update, init_func=init, frames=len(result_df), interval=100, repeat=False)
```

```python
from IPython.display import HTML
html = anim.to_html5_video()
HTML(html)
```
